```python
g="game"
import random 
def wrfile(board): 
    global g
    file = open ("D:\\"+g+".txt",'a')
    file.write(f"#____________________________Next Round________________________________#\n")
    file.write(f" +-------+-------+-------+\n")
    file.write(f" |   {board[(0,0)]}   |   {board[(0,1)]}   |   {board[(0,2)]}   |\n")
    file.write(f" +-------+-------+-------+\n")
    file.write(f" |   {board[(1,0)]}   |   {board[(1,1)]}   |   {board[(1,2)]}   |\n")
    file.write(f" +-------+-------+-------+\n")
    file.write(f" |   {board[(2,0)]}   |   {board[(2,1)]}   |   {board[(2,2)]}   |\n")
    file.write(f" +-------+-------+-------+\n")
    file.close()
#_________________________________________________________________________________________________#
def DisplayBoard(board): 
    print(f" +-------+-------+-------+")
    print(f" |   {board[(0,0)]}   |   {board[(0,1)]}   |   {board[(0,2)]}   |")
    print(f" +-------+-------+-------+")
    print(f" |   {board[(1,0)]}   |   {board[(1,1)]}   |   {board[(1,2)]}   |")
    print(f" +-------+-------+-------+")
    print(f" |   {board[(2,0)]}   |   {board[(2,1)]}   |   {board[(2,2)]}   |")
    print(f" +-------+-------+-------+")
    print(len(MakeListOfFreeFields(board)))


#_________________________________________________________________________________________________#
def MakeListOfFreeFields(board):
    '''The function browses the board and builds a list of all the free squares'''
    temp = {}
    for  i in board.items():
        if i[1]!="X" and i[1]!="O":
            temp [i[0]]=i[1]
    return temp
    
#__________________________________________________________________________________________________#
def EnterUserMove(board):
    '''Asks the user about their move, checks the input, and updates the board according to the user's decision'''
    move = int(input("Enter Your Move : "))
    if move in MakeListOfFreeFields(board).values():
        for i in  board.items():
            if i[1] == move:
                return i[0]
                break
    elif len(MakeListOfFreeFields(board))==0:
        print("Draw")
        return
    else:
        print("Cannt add Try Again")
        return EnterUserMove(board)
#__________________________________________________________________________________________________#
def DrawComputerMove(board):
    '''The function draws the computer's move and updates the board'''    
    f=True
    while f:
        p1=random.randint(0,2) 
        p2=random.randint(0,2) 
        p3=(p1,p2)
        for i in MakeListOfFreeFields(board).items():
            if(p3==i[0]):
                f=False       
    return p3     
#__________________________________________________________________________________________________# 
def VictoryFor(board):
    '''the function analyzes the board status in order to check 
    if the player using 'O'sor 'X's has won the game'''
    if (board[(0,0)]=="X" and board[(0,1)]=="X" and board[(0,2)]=="X") or (board[(1,0)]=="X" and board[(1,1)]=="X" and board[(1,2)]=="X") or (board[(2,0)]=="X" and board[(2,1)]=="X" and board[(2,2)]=="X") or  (board[(0,0)]=="X" and board[(1,0)]=="X" and board[(2,0)]=="X") or (board[(0,1)]=="X" and board[(1,1)]=="X" and board[(2,1)]=="X") or (board[(0,2)]=="X" and board[(1,2)]=="X" and board[(2,2)]=="X") or (board[(0,0)]=="X" and board[(1,1)]=="X" and board[(2,2)]=="X") or (board[(0,2)]=="X" and board[(1,1)]=="X" and board[(2,0)]=="X"):
        
        return 1
        
    elif(board[(0,0)]=="O" and board[(0,1)]=="O" and board[(0,2)]=="O") or (board[(1,0)]=="O" and board[(1,1)]=="O" and board[(1,2)]=="O") or (board[(2,0)]=="O" and board[(2,1)]=="O" and board[(2,2)]=="O") or  (board[(0,0)]=="O" and board[(1,0)]=="O" and board[(2,0)]=="O") or (board[(0,1)]=="O" and board[(1,1)]=="O" and board[(2,1)]=="O") or (board[(0,2)]=="O" and board[(1,2)]=="O" and board[(2,2)]=="O") or (board[(0,0)]=="O" and board[(1,1)]=="O" and board[(2,2)]=="O") or (board[(0,2)]=="O" and board[(1,1)]=="O" and board[(2,0)]=="O"):
        return 2
#__________________________________________________________________________________________________#
#playerConstant = 'O'
#computerConstant = 'X'

def Main():
    global g
    g+="1"
    
    board = {
    (0,0):1,
    (0,1):2,
    (0,2):3,
    (1,0):4,
    (1,1):5,
    (1,2):6,
    (2,0):7,
    (2,1):8,
    (2,2):9
     }
    print ( " im the computer and i  will start the game  :P " )
    for i in range (8):
      board [DrawComputerMove(board)]="X"
      if(VictoryFor(board)==1):
          print("pc win ")
          DisplayBoard(board)
          wrfile(board)
          break
      elif(VictoryFor(board) ==2):
          print("You win ")
          DisplayBoard(board)
          wrfile(board)
          break
      DisplayBoard(board)
      wrfile(board)
      board[EnterUserMove(board)]='O'
      if(VictoryFor(board)==1):
          print("pc win ")
          DisplayBoard(board)
          wrfile(board)
          break
      elif(VictoryFor(board) ==2):
          print("You win ")
          DisplayBoard(board)
          wrfile(board)
          break
#__________________________________________________________________________________________________#   
n=int(input("enter number of games:"))
for i in range(n):
    Main()

      



```
